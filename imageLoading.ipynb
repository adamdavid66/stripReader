```
from PIL import Image


#upload file
from google.colab import files
uploaded = files.upload()
for fn in uploaded.keys():
  fileuploaded = fn
 # we get the image and convert to greyscale
image = Image.open(fileuploaded).convert('LA')
image
```

```
#crop a section of the image
from IPython.display import display
cropped = image.crop((50,50,150,150))
cropped
```

```
import scipy
matrix = scipy.misc.imread(name=fileuploaded, flatten=True)
matrix
#scipy.cropped.imread()
```

```
import matplotlib.pyplot as plt

plt.plot(matrix)
plt.show()
```
